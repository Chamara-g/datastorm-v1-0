read sales_detal and create df_sales_detal

```
df_sales_detal = pd.read_csv(ds_path + 'sales_detail .csv')
```

create new col

```
years = df_sales_detal["year"].unique()
categories = df_sales_detal["catagory_code"].unique()
columns = ["customer_id"]

for year in years:
  for month in range(1, 13):
    for cat in categories:
      columns.append(str(year) + "_" + str(month) + "_" + cat)
```

new df create 

```
df_category = pd.DataFrame(columns=columns)
customer_ids = df_sales_detal['customer_id'].unique()
df_category.customer_id = customer_ids
df_category = df_category.set_index('customer_id')
```

```
start = time.process_time()

for index, row in df_sales_detal.iterrows():
  col_name = str(row.year) + "_" + str(row.month) + "_" + row.catagory_code
  df_category.xs(row.customer_id)[col_name] = row.total_spent_for_category


print(time.process_time() - start)
```

## *Processing sales summary*

```
df_sales_summary = pd.read_csv(ds_path + 'sales_summary .csv')
```

```
column_names = ['customer_id']

for year_value in ['2017', '2018', '2019']:
  for month_value in range(1,13):
    column_names.append(year_value + '_' + str(month_value) + '_' + 'weekend_amount_spent')
    column_names.append(year_value + '_' + str(month_value) + '_' + 'Num_of_weekend_visits')
    column_names.append(year_value + '_' + str(month_value) + '_' + 'total_amount_spent')
    column_names.append(year_value + '_' + str(month_value) + '_' + 'total_num_of_visits')

df_sales_summary_new = pd.DataFrame(columns = column_names)
```

```
customer_ids_sales_summary = df_sales_summary.customer_id.unique()
df_sales_summary_new.customer_id = customer_ids_sales_summary
df_sales_summary_new = df_sales_summary_new.set_index('customer_id')
```

```
start = time.process_time()

for index, row in df_sales_summary.iterrows():
  col_name1 = str(row.year) + "_" + str(row.month) + "_weekend_amount_spent"
  col_name2 = str(row.year) + "_" + str(row.month) + "_Num_of_weekend_visits"
  col_name3 = str(row.year) + "_" + str(row.month) + "_total_amount_spent"
  col_name4 = str(row.year) + "_" + str(row.month) + "_total_num_of_visits"
  df_sales_summary_new.xs(row.customer_id)[col_name1] = row.weekend_amount_spent
  df_sales_summary_new.xs(row.customer_id)[col_name2] = row.Num_of_weekend_visits
  df_sales_summary_new.xs(row.customer_id)[col_name3] = row.total_amount_spent
  df_sales_summary_new.xs(row.customer_id)[col_name4] = row.total_num_of_visits


print(time.process_time() - start)
```

```
df_sales_summary_new.to_csv(ds_write_path + 'df_sales_summary_new.csv', encoding='utf-8', mode='w', index=True)
```
